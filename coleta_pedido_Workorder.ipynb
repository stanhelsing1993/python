```python
import pandas as pd
import re
```

```python
# Substitua 'caminho/do/arquivo.xlsx' pelo caminho real do seu arquivo Excel
df = pd.read_csv('/content/WORK ORDER ABERTAS-data-27_07_2023, 11_55_30.csv', encoding= 'utf8')
#df = pd.read_excel('/content/Incident details.xlsx')
df.head()
```

```python
df_clone =  df[:]
```

```python
df_final = df_clone [['WorkorderID', 'Detailed_Description', 'Support_Group_Name']]
df_final.head()
```

```python
padrao_regex = r'\b4-\d{12}\b'
```

```python
def apply_regex(texto, regex_pattern):
    matches = re.findall(regex_pattern, texto)
    return ', '.join(matches) if matches else None

# Padrão regex
padrao_regex = r'\b4-\d{12}\b'

# Converte os valores da coluna 'X' para strings
df_final['Descrição'] = df_final['Detailed_Description'].astype(str)

# Aplica a função 'apply_regex' na coluna 'X' e salva os resultados na nova coluna 'Resultado'
df_final['pedido'] = df_final['Descrição'].apply(lambda texto: apply_regex(texto, padrao_regex))
```
